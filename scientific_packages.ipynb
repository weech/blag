```python
import numpy as np
```

```python
arr = np.linspace(-2*np.pi, 2*np.pi, 9)
arr = arr.reshape((3, 3))
print(arr)
```

```python
sined = np.sin(arr)
print(sined)
```

```python
sined[np.isclose(sined, 0)] = 0
print(sined)
```

```python
import matplotlib.pyplot as plt
from numpy import pi
# Create data
xs = np.linspace(-2*pi, 2*pi, 1000)
y_sin = np.sin(xs)
y_sin2 = np.sin(xs)**2
y_tan = np.tan(xs)

# Tangent goes off to infinity, so make coherent limit
ybottom, ytop = -2, 2
y_tan[np.logical_or(y_tan < ybottom, y_tan > ytop)] = np.nan

# Create plot
fig, ax0 = plt.subplots()
ax0.plot(xs, y_sin, color='red', label='sin(x)')
ax0.plot(xs, y_sin2, color='red', linestyle='dotted', label=r'$\sin{(x)}^2$') # LaTeX!
ax0.plot(xs, y_tan, color='blue', label='tan(x)')

# Prettify
ax0.legend()
ax0.set_title('An Assortment of Trig Functions')
ax0.set_xticks([n*pi/2 for n in range(-4, 5)])
ax0.set_ylim(bottom=ybottom, top=ytop)
plt.show()
```
